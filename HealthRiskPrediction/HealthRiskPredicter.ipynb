```python
from sklearn.preprocessing import OrdinalEncoder

ordinal_cols = ["diet", "stress", "alcohol"]
category_orders = [
    ["Poor", "Average", "Good"],    # diet
    ["Low", "Medium", "High"],      # stress
    ["Low", "Medium", "High"]       # alcohol
]

oe = OrdinalEncoder(categories=category_orders)

X_train[ordinal_cols] = oe.fit_transform(X_train[ordinal_cols])
X_test[ordinal_cols] = oe.transform(X_test[ordinal_cols])
```

```python
X_train
```

```python
from sklearn.preprocessing import OneHotEncoder
import pandas as pd

ohe_cols = ["smoking", "family_history"]
ohe = OneHotEncoder(sparse_output=False, drop="first", handle_unknown="ignore")

# Fit on train, transform both
train_encoded = ohe.fit_transform(X_train[ohe_cols])
test_encoded = ohe.transform(X_test[ohe_cols])

# Convert to DataFrame with new column names
ohe_train_df = pd.DataFrame(train_encoded, 
                            columns=ohe.get_feature_names_out(ohe_cols),
                            index=X_train.index)

ohe_test_df = pd.DataFrame(test_encoded, 
                           columns=ohe.get_feature_names_out(ohe_cols),
                           index=X_test.index)

# Drop original categorical columns and join new encoded ones
X_train = X_train.drop(columns=ohe_cols).join(ohe_train_df)
X_test = X_test.drop(columns=ohe_cols).join(ohe_test_df)
```

```python
# ordinal --> diet, stress, alcohol
# ohe --> smoking, family_history
# numerical --> age, sleep_hours, bmi, exercise_days
```

```python
from sklearn.preprocessing import StandardScaler
import pandas as pd

# Select numeric columns you want to scale
num_cols = ["age", "diet", "exercise_days", "bmi", "sleep_hours", "stress", "alcohol"]

scaler = StandardScaler()

# Fit on train, transform train
scaled_train = scaler.fit_transform(X_train[num_cols])
scaled_train_df = pd.DataFrame(scaled_train, columns=num_cols, index=X_train.index)

# Transform test (important: only transform, not fit again)
scaled_test = scaler.transform(X_test[num_cols])
scaled_test_df = pd.DataFrame(scaled_test, columns=num_cols, index=X_test.index)

print(scaled_train_df.head())
print(scaled_test_df.head())
```

```python
num_cols = ["age", "diet", "exercise_days", "bmi", "sleep_hours", "stress", "alcohol"]

# 1. Drop the numeric columns from original train/test
X_train_rest = X_train.drop(columns=num_cols)
X_test_rest  = X_test.drop(columns=num_cols)

# 2. Merge scaled numeric df with the remaining columns
X_train_final = pd.concat([X_train_rest, scaled_train_df], axis=1)
X_test_final  = pd.concat([X_test_rest, scaled_test_df], axis=1)

print(X_train_final.head())
print(X_test_final.head())
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

# 1. Initialize the model
log_reg = LogisticRegression(max_iter=1000, random_state=42)

# 2. Train (fit) the model on training data
log_reg.fit(X_train_final, y_train)

# 3. Predict on test data
y_pred = log_reg.predict(X_test_final)

# 4. Evaluate performance
print("Accuracy:", accuracy_score(y_test, y_pred))
print("\nConfusion Matrix:\n", confusion_matrix(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
```
